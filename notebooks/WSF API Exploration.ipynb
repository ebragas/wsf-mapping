```python
import requests
import gmplot
import pandas as pd
import os
```

```python
base_url = "http://www.wsdot.wa.gov/ferries/api/vessels/rest"
vessels_endpoint = "/vessellocations"
api_key = "eb0b9943-39c5-45cf-85d3-c38805037262"
```

```python
response = requests.get(base_url + vessels_endpoint, params={"apiaccesscode": api_key})
response.status_code
```

```python
response = requests.get("http://www.wsdot.wa.gov/ferries/api/vessels/rest/vessellocations?apiaccesscode=eb0b9943-39c5-45cf-85d3-c38805037262")
response.status_code
```

```python
data = response.json()
```

```python
for item in data:
    print({"lat": item["Latitude"], "lon": item["Longitude"]})
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
mega_map = gmplot.GoogleMapPlotter(47.604963, -122.439069, 16)
mega_map.plot(df["Latitude"], df["Longitude"], "blue", edge_width=1)
mega_map.draw("mega_map.html")
```
